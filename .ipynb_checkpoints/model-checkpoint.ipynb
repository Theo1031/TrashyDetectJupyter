```python
import torch
import torch.nn as nn
import torch.optim as optim
from torchvision import datasets, models, transforms
from torch.utils.data import DataLoader, random_split
import matplotlib.pyplot as plt
import numpy as np
import torchvision.transforms.functional as F
import torch.nn.functional as F
import os
import random



def train_and_evaluate(model, train_loader, validation_loader):
    criterion = nn.CrossEntropyLoss()
    optimizer = optim.SGD(model.parameters(), lr=0.001, momentum=0.9)

    num_epochs = 10  # Adjust as needed
    best_accuracy = 0.0

    for epoch in range(num_epochs):
        print("Starting to Train!")
        model.train()
        running_loss = 0.0
        for inputs, labels in train_loader:
            inputs, labels = inputs.to(device), labels.to(device)

            optimizer.zero_grad()
            outputs = model(inputs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item()

        print(f'Epoch {epoch+1}, Loss: {running_loss/len(train_loader)}')
        
        # Validation phase
        model.eval()
        correct = 0
        total = 0
        with torch.no_grad():
            for inputs, labels in validation_loader:  # Assuming you have a validation_loader
                inputs, labels = inputs.to(device), labels.to(device)
                outputs = model(inputs)
                _, predicted = torch.max(outputs.data, 1)
                total += labels.size(0)
                correct += (predicted == labels).sum().item()
        
        validation_accuracy = 100 * correct / total
        print(f'Epoch {epoch+1}, Validation Accuracy: {validation_accuracy}%')
        
        # Save model if it has better accuracy than any model saved before
        if validation_accuracy > best_accuracy:
            best_accuracy = validation_accuracy
            print(f'Saving new best model with accuracy: {best_accuracy}%')
            torch.save(model.state_dict(), 'model_best.pth')


def imshow(img, title=''):
    """Show image from tensor."""
    # If it's a tensor, convert to numpy
    if torch.is_tensor(img):
        img = img.cpu().numpy()
        img = np.transpose(img, (1, 2, 0))  # Change from (C, H, W) to (H, W, C)
        # Undo normalization
        mean = np.array([0.485, 0.456, 0.406])
        std = np.array([0.229, 0.224, 0.225])
        img = std * img + mean
        img = np.clip(img, 0, 1)  # Ensure the image values are between 0 and 1
    plt.imshow(img)
    plt.title(title)
    plt.show()


def load_model(model_path):
    # Make sure to define the exact same model architecture as was used during training
    model = models.resnet18(pretrained=False)  # Use pretrained=False since we're loading custom weights
    num_ftrs = model.fc.in_features
    model.fc = nn.Linear(num_ftrs, 2)  # Assuming your model outputs 2 classes, as before

    # Load the saved model state
    model.load_state_dict(torch.load(model_path, map_location=device))
    model.to(device)
    return model

def predict_image(image_path, model, threshold=0.51):
    # Load the image
    image = datasets.folder.default_loader(image_path)
    image = transform(image).unsqueeze(0).to(device)  # Apply transformations

    # Predict
    model.eval()
    with torch.no_grad():
        outputs = model(image)
        probabilities = F.softmax(outputs, dim=1)  # Get probabilities
        max_prob, preds = torch.max(probabilities, 1)  # Get the highest probability and predicted class
        class_confidence = max_prob.item()  # Convert to Python float for comparison

    # Decision
    prediction = 'neither' if class_confidence < threshold else train_data.classes[preds.item()]

    # Display the image and prediction
    plt.figure(figsize=(4, 4))
    plt.imshow(datasets.folder.default_loader(image_path))
    plt.title(f'Predicted: {prediction}, Confidence: {class_confidence:.2f}')
    plt.show()

def predict_and_display_first_10_images(test_dir, model):
    all_images = []
    
    # Get images from both cats and dogs directories
    categories = ['O', 'R']
    for category in categories:
        directory = os.path.join(test_dir, category)
        images = os.listdir(directory)
        
        # Append (image_path, category) to all_images list
        for image_name in images:
            image_path = os.path.join(directory, image_name)
            all_images.append((image_path, category))
    
    random.shuffle(all_images)
    selected_images = all_images[:10]
    for image_path, category in selected_images:
        print(f"Showing prediction for {category}: {os.path.basename(image_path)}")
        predict_image(image_path, model)

if __name__ == '__main__':
    transform = transforms.Compose([
        transforms.Resize(256),
        transforms.CenterCrop(224),
        transforms.ToTensor(),
        transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
    ])

    train_data = datasets.ImageFolder(root='DATASET/TRAIN', transform=transform)
    test_data = datasets.ImageFolder(root='DATASET/TEST', transform=transform)

    validation_size = int(0.15 * len(train_data))  #15% of the training set for validation

    training_size = len(train_data) - validation_size
    train_dataset, validation_dataset = random_split(train_data, [training_size, validation_size])

    train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True, num_workers=2)
    validation_loader = DataLoader(validation_dataset, batch_size=32, shuffle=False, num_workers=2)
    test_loader = DataLoader(test_data, batch_size=32, shuffle=False, num_workers=2)


    model = models.resnet18(pretrained=True)
    num_ftrs = model.fc.in_features
    model.fc = nn.Linear(num_ftrs, 2)  # 2 classes: cats and dogs

    device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
    model.to(device)

    #train_and_evaluate(model, train_loader, validation_loader)

    best_model = load_model('model_best.pth')

    test_cats_dir = 'DATASET/TEST/O'
    test_dogs_dir = 'DATASET/TEST/R'
    test_dir = 'DATASET/TEST'
```
